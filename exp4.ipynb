```python
#1 Write a program to count and display the number of capital letters in a given string.
input_string = input("Enter a string: ")
capital_count = 0
for char in input_string:
  if char.isupper():
    capital_count += 1
print(f"The number of capital letters in the string is: {capital_count}")
```

```python
#2 Count total number of vowels in a given string.
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in string:
  if char in vowels:
    count += 1
print(f"The total number of vowels in the string is: {count}")
```

```python

#4 WAP to enter a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left.
string = input("Enter a string: ")
substring = input("Enter a substring: ")
c=0
for i in range( len(string)- len(substring)+1):
  if string[i:i+len(substring)] == substring:
    c+=1
print("The number of times the substring occurs in the given string is: ",c)

```

```python
#5 Given a string containing both upper and lower case alphabets. Write a Python program to count the number of occurrences of each alphabet (case insensitive) and display the same.
string = input("Enter a string:")
lower_string = string.lower()
c={ }
for char in lower_string:
  if 'a' <= char <= 'z':
    if char in c:
      c[char] += 1
    else:
      c[char] = 1
print("Alphabet occurrences:")
for char, count in c.items():
  print(f"{char}: {count}")
```

```python
#6 Program to count number of unique words in a given sentence using sets.
sentence = input("Enter a sentence: ")
words_list = sentence.lower().split()
unique_words_set = set(words_list)
count = len(unique_words_set)

print(f"The sentence is: {sentence}")
print(f"The unique words are: {unique_words_set}")
print(f"Number of unique words: {count}")
count_concise = len(set(sentence.lower().split()))
print(f"Concise count: {count_concise}")
```

```python
#7 Create 2 sets s1 and s2 of n fruits each by taking input from user and find:
# a) Fruits which are in both sets s1 and s2
# b) Fruits only in s1 but not in s2
# c) Count of all fruits from s1 and s2
s1 = set(input("Enter the fruits in set 1: ").split())
s2 = set(input("Enter the fruits in set 2: ").split())
print("Fruits in both sets: ", s1.intersection(s2))
print("Fruits in set 1 but not in set 2: ", s1.difference(s2))
print("Fruits in set 2 but not in set 1: ", s2.difference(s1))
print("Count of all fruits from set 1: ", len(s1))
print("Count of all fruits from set 2: ", len(s2))
```

```python
#8 Take two sets and apply various set operations on them :
# S1 = {Red ,yellow, orange , blue }
# S2 = {violet, blue , purple}
S1 = { 'Red' ,'yellow', 'orange' , 'blue' }
S2 = { 'violet', 'blue' , 'purple' }
print("Union of S1 and S2: ", S1.union(S2))
print("Intersection of S1 and S2: ", S1.intersection(S2))
print("Difference of S1 and S2: ", S1.difference(S2))
print("Symmetric difference of S1 and S2: ", S1.symmetric_difference(S2))
```
