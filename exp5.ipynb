```python
#1 Scan n values in range 0-3 and print the number of times each value has occurred.
def count_occurrences_manual(data_sequence):
    print(f"Counting occurrences in: {data_sequence}")
    for value in range(4):
        count = data_sequence.count(value)
        print(f"Value {value} occurred {count} times")
list_values = [0, 1, 2, 2, 3, 1, 0, 3, 2, 1]
count_occurrences_manual(list_values)
tuple_values = (0, 1, 2, 2, 3, 1, 0, 3, 2, 1)
count_occurrences_manual(tuple_values)
```

```python
#4  Create a dictionary of n persons where key is name and value is city.
# a) Display all names
# b) Display all city names
# c) Display student name and city of all students.
# d) Count number of students in each city.
# Create a dictionary of n persons

n = int(input("Enter number of persons: "))
person_city = {}
for _ in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    person_city[name] = city
print("\n--- All Names ---")
print(list(person_city.keys()))
print("\n--- All City Names ---")
print(list(person_city.values()))
print("\n--- Student Details (Name: City) ---")
for name, city in person_city.items():
    print(f"{name}: {city}")
print("\n--- Student Count per City ---")
city_counts = {}
for city in person_city.values():
    city_counts[city] = city_counts.get(city, 0) + 1

for city, count in city_counts.items():
    print(f"{city}: {count}")
```

```python
#5 Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name, year, director name, production cost, collection made (earning) & perform the following :-
# a) print all movie details
# b) display name of movies released before 2015
# c) print movies that made a profit.
# d) print movies directed by a particular director

def movie_database():
    movies = {}
    n = int(input("Enter the number of movies: "))
    for i in range(n):
        print(f"\nEnter details for movie {i+1}:")
        name = input("Name: ")
        year = int(input("Year: "))
        director = input("Director: ")
        cost = float(input("Production Cost: "))
        collection = float(input("Collection Made: "))
        movies[name] = {
            "year": year,
            "director": director,
            "cost": cost,
            "collection": collection
        }
    print("\n--- All Movie Details ---")
    for name, details in movies.items():
        print(f"Movie: {name}, Year: {details['year']}, "
              f"Director: {details['director']}, Cost: {details['cost']}, "
              f"Collection: {details['collection']}")
    print("\n--- Movies Released Before 2015 ---")
    before_2015 = [name for name, d in movies.items() if d['year'] < 2015]
    print(before_2015 if before_2015 else "None")
    print("\n--- Profitable Movies ---")
    profitable = [name for name, d in movies.items() if d['collection'] > d['cost']]
    print(profitable if profitable else "None")
    print("\n--- Search by Director ---")
    search_dir = input("Enter director name: ")
    dir_movies = [name for name, d in movies.items() if d['director'] == search_dir]
    print(f"Movies by {search_dir}: {dir_movies}" if dir_movies else "No movies found.")
movie_database()
```

```python
#6 Create a contact book where users can store, search, update, and delete contacts. Use dictionary for storing contacts.

contacts = {}
def add_contact(name, phone, email):
    """Adds a new contact, checking for duplicates."""
    if name in contacts:
        print(f"Error: Contact '{name}' already exists.")
    else:
        contacts[name] = {'phone': phone, 'email': email}
        print(f"Contact '{name}' added successfully.")

def view_contacts():
    """Displays all saved contacts."""
    if not contacts:
        print("Your contact book is empty.")
    else:
        print("\n--- Your Contacts ---")
        for name, details in contacts.items():
            print(f"Name: {name}, Phone: {details['phone']}, Email: {details['email']}")

def update_contact(name, phone=None, email=None):
    """Updates an existing contact."""
    if name in contacts:
        if phone: contacts[name]['phone'] = phone
        if email: contacts[name]['email'] = email
        print(f"Contact '{name}' updated successfully.")
    else:
        print(f"Error: Contact '{name}' not found.")

def delete_contact(name):
    """Deletes a contact."""
    if name in contacts:
        del contacts[name]
        print(f"Contact '{name}' deleted successfully.")
    else:
        print(f"Error: Contact '{name}' not found.")

def menu():
    """Menu-driven interface."""
    while True:
        print("\n1. Add, 2. View, 3. Update, 4. Delete, 5. Exit")
        choice = input("Enter choice: ")
        if choice == '1':
            name = input("Name: ")
            add_contact(name, input("Phone: "), input("Email: "))
        elif choice == '2': view_contacts()
        elif choice == '3': update_contact(input("Name: "), input("New Phone: "), input("New Email: "))
        elif choice == '4': delete_contact(input("Name: "))
        elif choice == '5': break

if __name__ == "__main__":
    menu()
```

```python
#7 Create a Todo list Manager where users can add, view, and remove tasks. Use List for storing tasks.

tasks = []

def add_task():
    task = input("Enter a new task: ")
    if task: tasks.append(task)

def view_tasks():
    for index, task in enumerate(tasks, start=1):
        print(f"{index}. {task}")

def remove_task():
    view_tasks()
    idx = int(input("Enter number to remove: ")) - 1
    if 0 <= idx < len(tasks): tasks.pop(idx)

# Main loop with menu options
while True:
    choice = input("1.Add, 2.View, 3.Remove, 4.Exit: ")
    if choice == '1': add_task()
    elif choice == '2': view_tasks()
    elif choice == '3': remove_task()
    elif choice == '4': break
```
